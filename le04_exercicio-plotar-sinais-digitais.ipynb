# Sinais discretos
Desafio: Plote os seguintes sinais discretos

- $ y_1(t) = 2 \sin(2\pi f t) $
- $ y_2(t) = \sin(2\pi f t) + \sin(4\pi f t) $
- $ y_3(t) = \frac{4}{\pi} \sin(2\pi f t) + \frac{4}{3\pi} \sin(6\pi f t) $
- $ y_4(t) = y_3(t) + \frac{4}{5\pi} \sin(10\pi f t) + \frac{4}{7\pi} \sin(14\pi f t) + \frac{4}{9\pi} \sin(18\pi f t) $

# ![sines.png](sines.png)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#frequncia
f = 2.0

#tempo
t = np.linspace(0, 1, 1000)

y1 = 2*np.sin(2*np.pi*f*t)

y2 = np.sin(2*np.pi*f*t)+np.sin(4*np.pi*f*t)

fig, ax = plt.subplots()
ax.plot(t, y1)
plt.show()

fig, ax = plt.subplots()
ax.plot(t, y2)
plt.show()
```

```python
#tempo
t = np.linspace(0, 1, 1000)

#frequncia
f = 2.0


y3 = 4/np.pi*np.sin(2*np.pi*f*t)+ 4/3*np.pi*np.sin(6*np.pi*f*t)

y4 = y3 + 4/5*np.pi * np.sin(10*np.pi*f*t) + 4/7*np.pi * np.sin(14*np.pi*f*t) +4/9*np.pi * np.sin(18*np.pi*f*t)


fig, ax = plt.subplots()
ax.plot(t, y3)
plt.show()

fig, ax = plt.subplots()
ax.plot(t, y4)
plt.show()
```

```python
#transformada de fourier (DFT)
 
def dft(x):
    
    N = len(x)
    X = np.zeros(N, dtype=complex)

    for k in range(N):
        for n in range(N):
            X[k] += x[n]*np.exp(-1j * 2 * np.pi * n * k / N)
    
    return X
t = np.linspace(0, 0.1, 1000)
f =  40
y1 = 2*np.sin(2*np.pi*f*t)
```

```python

plt.plot(t, y1)
```

```python
Y1 = dft(y1)
```
