# Aliasing

```python
fs = 40  
ts = 1/fs
f0 = 2 #hertz

n = np.arange(fs)

xn = np.cos(2* np.pi * f0 * n * ts)

#aliasing
k = 1 
df = k*fs
x2n = np.cos(2* np.pi * (f0+df) * n * ts)

plt.plot(n*ts, xn, label="Sinal Amostrado")
plt.plot(n*ts, x2n,'o', label="Sinal com Aliasing",)
plt.title("Sinal amostral")
plt.legend()
plt.ylabel("Magnitude")
plt.xlabel("Tempo (s)")
```

```python
#Existem mais de um sinal 
#contínuo que corresponde ao sinal discreto

fs2 = 400
ts2 = 1/fs2
n2 = np.arange(fs2)
#xfs2n = np.cos(2 * np.pi * (f0+df) *n2 * ts2)
xfs2n = np.cos(2 * np.pi * (-f0+df) *n2 * ts2)
```

```python
plt.plot(n*ts, xn, label="Sinal Amostrado")
plt.plot(n*ts, x2n,'o', label="Sinal com Aliasing",)
plt.plot(n2*ts2, xfs2n, label="Sinal com frequencia alta")
plt.title("Sinal amostral")
plt.legend()
plt.ylabel("Magnitude")
plt.xlabel("Tempo (s)")
```
