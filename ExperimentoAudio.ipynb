```python
2#transformada de fourier (DFT)
import numpy as np
import matplotlib.pyplot as plt

 
def dft(x):
    
    N = len(x)
    X = np.zeros(N, dtype=complex)

    for k in range(N):
        for n in range(N):
            X[k] += x[n]*np.exp(-1j * 2 * np.pi * n * k / N)
    
    return X
```

```python
X = dft(x[:1000])

np.fft.fft(x)
```

```python
import matplotlib.pyplot as plt
Xmag = np.abs(X)
plt.plot(Xmag)
```

```python
def dft1(x):
    
    N = len(x)
    n = np.arange(N)
    
    X = np.zeros(N, dtype=complex)
    
    
    for k in np.arange(N):
        w = np.exp(-1j * (2 * np.pi / N) * k * n)
        
        #X[k] += np.inner(w.T, x)
        # ou 
        X[k] += w.T @ x
    
    return X
```

```python
Y1 = dft1(x[:1000])
```

```python
import matplotlib.pyplot as plt
Y1mag = np.abs(Y1)
plt.plot(Y1mag)
```

```python
def dft2(x):
      
    N = len(x)

    n = np.expand_dims(np.arange(N), 1)
    
    W = np.exp(-1j * (2 * np.pi / N) * (n @ n.T))
        
    X =  W @ x
    
    return X
```

```python
Y2 = dft2(x[:1000])
```

```python
import matplotlib.pyplot as plt
Y2mag = np.abs(Y2)
plt.plot(Y2mag)
```
